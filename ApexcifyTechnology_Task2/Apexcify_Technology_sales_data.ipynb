***Loading data using pandas***

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_data_sample.csv', encoding='latin1')

# Group by year and month, sum SALES
monthly = (df.groupby(['YEAR_ID', 'MONTH_ID'])['SALES']
             .sum()
             .reset_index()
             .sort_values(['YEAR_ID', 'MONTH_ID']))

print(monthly)
```

***plot the data using the matplot library***

```python
# Plot monthly sales trend per year
months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']

plt.figure(figsize=(12, 5))
for year, group in monthly.groupby('YEAR_ID'):
    plt.plot(group['MONTH_ID'], group['SALES'],
             marker='o', linewidth=2, label=str(year))

plt.xlabel('Month')
plt.ylabel('Sales (USD)')
plt.title('Monthly Sales Trend â 2003 to 2005')
plt.xticks(range(1,13), months)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
