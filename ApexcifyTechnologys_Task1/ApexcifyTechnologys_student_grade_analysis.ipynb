***TASK 1: Analyze Student Grades***
**Objective:**
- Load a CSV file containing student names and scores in 3 subjects.
- Calculate each studentâ€TMs average score.
- Find the student with the highest average.
Tools:
- Python, Pandas
Sample Columns:
Name | Math | English | Science
Key Concepts:
- DataFrames, mean(), max()

**- Load a CSV file containing student names and scores in 3 subjects.**

```python
import pandas as pd

data = pd.read_csv('./students.csv')

df = pd.DataFrame(data)
```

**- Calculate each studentâ€TMs average score.**

```python
df["Average"] = df[["Math", "English", "Science"]].mean(axis=1)

print(df)
```

**- Find the student with the highest average.**

```python
top_student = df.loc[df["Average"].idxmax()]

print("Top Student:")
print(top_student)
```
